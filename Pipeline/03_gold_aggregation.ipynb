Defects_user_lab_gold

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.defects_user_lab_gold AS
SELECT
    a.user,
    COALESCE(u.diplayname, l.diplayname) AS diplayname,
    COALESCE(u.department, l.department) AS department,
    COALESCE(u.mail, l.mail) AS mail,
    COALESCE(u.samdomain_name, l.samdomain_name) AS samdomain_name,
    COALESCE(u.lab_name, l.lab_name) AS lab_name,
    COALESCE(u.region, l.region) AS region,
    COALESCE(u.lab_source, l.lab_source) AS lab_source,
    a.data_provider,
    a.month,
    a.event_type,
    a.start_datetime,
    a.week_day,
    a.day,
    a.object,
    a.hour,
    a.year,
    a.event_duration_seconds,
    a.top_level_folder
FROM workspace.silver.audit_defects_silver a
LEFT JOIN workspace.silver.user_details_silver u
    ON a.user = u.username
LEFT JOIN workspace.silver.ldap_users_silver l
    ON a.user = l.username;
```

```python
%sql
SELECT *
FROM workspace.gold.defects_user_lab_gold
WHERE lab_name IS NULL;
```

All Applications

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.user_lab_application_gold AS
SELECT
    a.user,
    a.application_name,
    a.data_provider,
    a.month,
    a.year,
    a.week_day,
    a.day,
    a.hour,
    a.event_type,
    a.start_datetime,
    a.event_duration_seconds,
    a.object,
    a.top_level_folder,
    COALESCE(u.diplayname, l.diplayname) AS diplayname,
    COALESCE(u.department, l.department) AS department,
    COALESCE(u.mail, l.mail) AS mail,
    COALESCE(u.lab_name, l.lab_name) AS lab_name,
    COALESCE(u.region, l.region) AS region,
    COALESCE(u.lab_source, l.lab_source) AS lab_source
FROM workspace.silver.audit_defects_silver a
LEFT JOIN workspace.silver.user_details_silver u
    ON a.user = u.username
LEFT JOIN workspace.silver.ldap_users_silver l
    ON a.user = l.username
```

```python
%sql
select * from workspace.gold.user_lab_application_gold;
```

Application_user_lab_summary

```python
%sql
CREATE OR REPLACE TABLE workspace.gold.user_lab_application_summary_gold AS
SELECT
    application_name,
    user,
    diplayname,
    department,
    mail,
    lab_name,
    region,
    lab_source,
    COUNT(*) AS usage_count,
    MIN(start_datetime) AS first_usage,
    MAX(start_datetime) AS last_usage
FROM workspace.gold.user_lab_application_gold
GROUP BY
    application_name,
    user,
    diplayname,
    department,
    mail,
    lab_name,
    region,
    lab_source;
```
